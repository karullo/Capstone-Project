```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

survey_data = pd.read_csv('/content/drive/MyDrive/Python Notebooks/survey_results_public.csv')
survey_data
```

```python
# Filter Country column to only USA respondents

survey_data_usa = survey_data.loc[survey_data['Country'] == 'United States of America']

survey_data_usa
```

```python
#Remove incomplete salary data

survey_data_usa = survey_data_usa.dropna(subset=['ConvertedCompYearly'])

survey_data_usa
```

```python
#Filter for full-time employees: Keep only Employment = 'Employed'

survey_data_usa = survey_data_usa.loc[survey_data_usa['Employment'] == 'Employed, full-time']
survey_data_usa
```

```python
# Filter salaries between $30,000 and $500,000

survey_data_usa = survey_data_usa.loc[(survey_data_usa['ConvertedCompYearly'] >= 30000) & (survey_data_usa['ConvertedCompYearly'] <= 500000)]
survey_data_usa
```

```python
#Filter experience: Keep YearsCodePro between 1 and 30 years

survey_data_usa['YearsCodePro'] = pd.to_numeric(survey_data_usa['YearsCodePro'], errors='coerce')

survey_data_usa = survey_data_usa.dropna(subset=['YearsCodePro'])

survey_data_usa = survey_data_usa.loc[(survey_data_usa['YearsCodePro'] >= 1) & (survey_data_usa['YearsCodePro'] <= 30)]
survey_data_usa
```

```python
survey_data_usa.to_csv('survey_USA_cleaned.csv')
```

```python
survey_data_gender = survey_data_usa[['Gender', 'ConvertedCompYearly', 'YearsCodePro', 'DevType']]
survey_data_gender
```

```python
# Filter to respondents who identified as “Man” or “Woman”

survey_data_gender = survey_data_gender[survey_data_gender['Gender'].isin(['Man', 'Woman'])]

survey_data_gender
```

```python
# filter by experience (5-10 years)

survey_data_gender= survey_data_gender[(survey_data_gender['YearsCodePro'] >= 5) & (survey_data_gender['YearsCodePro'] <= 10)]
survey_data_gender
```

```python
# t-test of men and women salary

from scipy import stats

t_statistic, p_value = stats.ttest_ind(df_men['ConvertedCompYearly'], df_women['ConvertedCompYearly'])

print("T-statistic:", t_statistic)
print("P-value:", p_value)

```

```python
# z-test for proportion of genders

import numpy as np
from statsmodels.stats.proportion import proportions_ztest


count_women = len(df_women)

total_sample_size = len(survey_data_gender)

p_0 = 0.184


z_statistic, p_value = proportions_ztest(count_women, total_sample_size, p_0)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)
```

```python
# z-test for proportion based on population (18.7)


count_women = len(df_women)

total_sample_size = len(survey_data_gender)

p_0 = 0.5


z_statistic, p_value = proportions_ztest(count_women, total_sample_size, p_0)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)
```
